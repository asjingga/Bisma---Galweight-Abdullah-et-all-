```python
#Jika menggunakan Aturan Silverman

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px

# Fungsi Gaussian kernel
def gaus(t):
    return np.exp(-(t**2)/2) / np.sqrt(2 * np.pi)

# Perhitungan bandwidth standar dengan aturan Silverman
band = 1.06 * df1['zspek'].std() * (len(df1)**(-1/5))

# Aturan Scott untuk bandwidth
#band = 3.49 * df1['zspek'].std() * (len(df1)**(-1/3))

# Pilot Density Estimation (f00)
def f00(x1, band):
    B = 0
    for j in range(len(df1)):
        X = (x1 - df1.at[j, 'zspek']) / band
        gauss = gaus(X)
        B += gauss
    return B / (len(df1) * band)

# Perhitungan log-likelihood untuk gamma
def calculate_gamma(df1, band):
    log_sum = 0
    for i in range(len(df1)):
        B = 0
        for j in range(len(df1)):
            X = (df1.at[i, 'zspek'] - df1.at[j, 'zspek']) / band
            gauss = gaus(X)
            B += gauss
        hasil = B / (len(df1) * band)
        log_sum += np.log10(hasil)
    return np.power(10, log_sum / len(df1))

# Hitung gamma sekali untuk keseluruhan dataset
gamma = calculate_gamma(df1, band)

# Fungsi AKM dengan bandwidth adaptif
def adaptive_kernel_method(x2, df1, band, gamma):
    C = 0
    for k in range(len(df1)):
        f00_k = f00(df1.at[k, 'zspek'], band)
        adaptive_bandwidth = band * np.sqrt(gamma / f00_k)
        W = (x2 - df1.at[k, 'zspek']) / adaptive_bandwidth
        gauss = gaus(W) / adaptive_bandwidth
        C += gauss
    return C / len(df1)

# Membuat range nilai z untuk diprediksi
test = np.linspace(0.074, 0.086, 200)

# Hitung prediksi z menggunakan fungsi adaptive kernel method
zpred = np.array([adaptive_kernel_method(t, df1, band, gamma) for t in test])

# Plot histogram dengan Matplotlib dan hasil prediksi dengan adaptive kernel method
plt.subplots(figsize=(12, 6))

plt.hist(df1['zspek'], bins=26, alpha=0.8,
         histtype='bar', color='blue',
         edgecolor='black', density=True)
plt.plot(test, zpred, color='red')
plt.xlim(0.074, 0.086)
plt.xlabel('z')
plt.ylabel('Density')
plt.title('Histogram of redshift with 1D AKM with Silverman Method')
plt.show()

# Plot menggunakan Plotly
a = pd.DataFrame(dict(
    z=test,
    n=zpred
))

fig = px.line(a, x='z', y='n', title='Adaptive Kernel Density Estimation with Plotly')
fig.show()
```

```python
#Jika menggunakan Aturan Sheather-Jones

import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px

# Fungsi Gaussian kernel
def gaus(t):
    return np.exp(-(t**2)/2) / np.sqrt(2 * np.pi)

# Gunakan statsmodels untuk menghitung bandwidth menggunakan Sheather-Jones
kde = sm.nonparametric.KDEUnivariate(df1['zspek'])
kde.fit(kernel='gau', bw='scott')  # Dapat diganti dengan 'scott', 'silverman', atau 'normal_reference'

# Bandwidth Sheather-Jones yang dihitung oleh statsmodels
stjj_sj = kde.bw

# Pilot Density Estimation (f00) menggunakan stjj_sj
def f00_sj(x1, stjj_sj):
    B = 0
    for j in range(len(df1)):
        X = (x1 - df1.at[j, 'zspek']) / stjj_sj
        gauss = gaus(X)
        B += gauss
    return B / (len(df1) * stjj_sj)

# Perhitungan log-likelihood untuk gamma menggunakan stjj_sj
def calculate_gamma_sj(df1, stjj_sj):
    log_sum = 0
    for i in range(len(df1)):
        B = 0
        for j in range(len(df1)):
            X = (df1.at[i, 'zspek'] - df1.at[j, 'zspek']) / stjj_sj
            gauss = gaus(X)
            B += gauss
        hasil = B / (len(df1) * stjj_sj)
        log_sum += np.log10(hasil)
    return np.power(10, log_sum / len(df1))

# Hitung gamma sekali untuk keseluruhan dataset menggunakan stjj_sj
gamma_sj = calculate_gamma_sj(df1, stjj_sj)

# Fungsi AKM dengan bandwidth adaptif menggunakan stjj_sj
def adaptive_kernel_method_sj(x2, df1, stjj_sj, gamma_sj):
    C = 0
    for k in range(len(df1)):
        f00_k = f00_sj(df1.at[k, 'zspek'], stjj_sj)
        adaptive_bandwidth = stjj_sj * np.sqrt(gamma_sj / f00_k)
        W = (x2 - df1.at[k, 'zspek']) / adaptive_bandwidth
        gauss = gaus(W) / adaptive_bandwidth
        C += gauss
    return C / len(df1)

# Membuat range nilai z untuk diprediksi
test = np.linspace(0.074, 0.086, 150)

# Hitung prediksi z menggunakan fungsi adaptive kernel method dengan Sheather-Jones bandwidth
zpred_sj = np.array([adaptive_kernel_method_sj(t, df1, stjj_sj, gamma_sj) for t in test])

# Plot histogram dengan Matplotlib dan hasil prediksi dengan adaptive kernel method Sheather-Jones
plt.subplots(figsize=(12, 6))

plt.hist(df1['zspek'], bins=26, alpha=0.8,
         histtype='bar', color='blue',
         edgecolor='black', density=True)
plt.plot(test, zpred_sj, color='red')
plt.xlim(0.074, 0.086)
plt.xlabel('z')
plt.ylabel('Density')
plt.title('Histogram of redshift with 1D AKM Prediction (Sheather-Jones)')
plt.show()

# Plot menggunakan Plotly
a_sj = pd.DataFrame(dict(
    z=test,
    n=zpred_sj
))

fig_sj = px.line(a_sj, x='z', y='n', title='Adaptive Kernel Density Estimation (Sheather-Jones) with Plotly')
fig_sj.show()
```

#Perhitungan jarak proyeksi

```python
def jarak (ra,rap,dec,decp):
  delra=np.abs(ra-rap)
  cos = np.cos(np.radians(90-dec))*np.cos(np.radians(90-decp))
  sin = np.sin(np.radians(90-dec))*np.sin(np.radians(90-decp))*np.cos(np.radians(delra))
  C = sin + cos
  ar = np.arccos(C)
  return np.tan(ar)*dl
```

```python
import numpy as np
import matplotlib.pyplot as plt

def get_figcontours(figs, xi, yi, lev=None, plot='Yes'):
    # Checking if `allsegs` is available, otherwise using `get_paths()`
    try:
        # Matplotlib versions before 3.8
        lines = figs.allsegs
    except AttributeError:
        # Matplotlib versions 3.8 and above
        lines = [path.vertices for path in figs.collections]

    contour_all = []
    for level_lines in lines:
        contour_each = []
        if isinstance(level_lines, list):  # For Matplotlib versions < 3.8
            for segment in level_lines:
                for point in segment:
                    contour_each.append(list(point))
        else:  # For Matplotlib versions >= 3.8
            contour_each = [list(point) for point in level_lines]

        contour_all.append(contour_each)

    if plot == 'Yes':
        # Plot individual contour lines
        plt.figure(figsize=(17, 17))
        for i, contour in enumerate(contour_all):
            plt.subplot(6, 4, i + 1)
            contour = np.array(contour)
            plt.plot(contour[:, 0], contour[:, 1], '.')
            if lev is not None:
                plt.title('Contour level {}'.format(np.round(lev[i] / 100, 1)))
            plt.xlim(np.min(xi), np.max(xi))
            plt.ylim(np.min(yi), np.max(yi))
        plt.show()

    return contour_all

def compute_contour_area(contour_all):
    # Using Green's Theorem to calculate the area enclosed by each contour
    def area(vertices):
        a = 0
        x0, y0 = vertices[0]
        for x1, y1 in vertices[1:]:
            dx = x1 - x0
            dy = y1 - y0
            a += 0.5 * (y0 * dx - x0 * dy)
            x0, y0 = x1, y1
        return abs(a)

    return [area(np.array(contour)) for contour in contour_all]

# Example usage:
# contour_all = get_figcontours(CS, RR, VVV, lev=lev, plot='Yes')
# contour_area = compute_contour_area(contour_all)
```

```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
from astropy.coordinates import SkyCoord
import astropy.units as u

# Fungsi untuk menghitung jarak antar galaksi
def hitung_jarak(ra1, dec1, ra2, dec2):
    # Membuat koordinat galaksi 1 dan galaksi 2 dalam satuan derajat
    coord1 = SkyCoord(ra=ra1*u.degree, dec=dec1*u.degree)
    coord2 = SkyCoord(ra=ra2*u.degree, dec=dec2*u.degree)

    # Menghitung jarak sudut antara dua galaksi
    return coord1.separation(coord2).deg  # Jarak dalam derajat

#Array kosong untuk menyimpan jarak antar galaksi
R_ij = np.empty((len(A0279), len(A0279)), dtype=float)

#Mencari jarak setiap galaksi dengan galaksi anggota lain berdasarkan posisi galaksi
for i in range(len(A0279)):
    for j in range(len(A0279)):
        R_ij[i][j] = hitung_jarak(A0279['ra'].iloc[i], A0279['dec'].iloc[i],
                                          A0279['ra'].iloc[j], A0279['dec'].iloc[j])

# Ubah array jarak antar galaksi ke DataFrame dan jadikan tipe float
jarakantargalaksi = pd.DataFrame(R_ij)
jarakantargalaksi_f = jarakantargalaksi.astype(float)

#Cari 10 tetangga terdekat untuk setiap galaksi
tetanggaterdekat = np.empty(shape=(len(A0279), 11), dtype='float')

for i in range(len(A0279)):
    # Cari 11 nilai terkecil termasuk jarak ke galaksi itu sendiri
    data = jarakantargalaksi_f.iloc[i].nsmallest(11).values
    # Masukkan ke array tetanggaterdekat
    tetanggaterdekat[i] = data

# Ubah array tetangga terdekat ke DataFrame dan ubah ke tipe float
tetangga = pd.DataFrame(tetanggaterdekat).astype(float)

# 4. Hitung kerapatan proyeksi (rho) dengan logaritma
rho = []
for i in range(len(A0279)):
    # Jarak ke tetangga terjauh
    logi = math.log10(10 / (np.pi * (tetangga.iloc[i, 10] ** 2)))
    rho.append(logi)

# Simpan hasil kerapatan ke dalam DataFrame A0279
A0279['rho'] = pd.DataFrame(rho)
A0279 = A0279.reset_index(drop=True)
```

# Penentuan Massa Heisler

```python
import numpy as np
import pandas as pd

# Fungsi untuk menghitung jarak sudut dalam derajat
def hitung_jarak_angular(ra1, dec1, ra2, dec2):
    delra = np.abs(ra1 - ra2)
    cos = np.cos(np.radians(90 - dec1)) * np.cos(np.radians(90 - dec2))
    sin = np.sin(np.radians(90 - dec1)) * np.sin(np.radians(90 - dec2)) * np.cos(np.radians(delra))
    C = sin + cos
    ar = np.arccos(C)
    return np.degrees(ar)  # Jarak dalam derajat

# Asumsikan `A0279` adalah DataFrame yang berisi data galaksi
# Misalnya: A0279 = pd.DataFrame({'ra': [...], 'dec': [...]})

# Tentukan pusat gugus galaksi (misalnya: rata-rata RA dan Dec dari gugus galaksi atau nilai tertentu)
center_ra = 29.070641492322501
center_dec = 1.050816667387900

# Definisikan jarak angular diameter distance (D_A) dalam Mpc
D_A = 354.61195  #Mpc

# Membuat array untuk menyimpan jarak antara pusat dan galaksi lain dalam Mpc
R_center_to_galaxy_Mpc = []

# Menghitung jarak antara pusat gugus dengan setiap galaksi dan mengonversi ke Mpc
for i in range(len(A0279)):
    ra_galaxy = A0279['ra'].iloc[i]
    dec_galaxy = A0279['dec'].iloc[i]
    # Hitung jarak sudut dalam derajat
    jarak_angular = hitung_jarak_angular(center_ra, center_dec, ra_galaxy, dec_galaxy)
    # Konversi jarak sudut dari derajat ke radian
    jarak_radian = np.radians(jarak_angular)
    # Konversi jarak sudut ke Mpc menggunakan D_A
    jarak_Mpc = D_A * jarak_radian
    R_center_to_galaxy_Mpc.append(jarak_Mpc)

# Ubah list ke dalam DataFrame
jarak_pusat_galaksi_f = pd.DataFrame(R_center_to_galaxy_Mpc, columns=['Jarak ti Pusat'])

# Mengonversi DataFrame ke tipe float jika diperlukan
jarak_pusat_galaksi_f = jarak_pusat_galaksi_f.astype(float)

# Menampilkan DataFrame
print(jarak_pusat_galaksi_f)

```
